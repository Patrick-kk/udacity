使用__Python__读取__CSV__文件

以下为读取方法：

````python
import unicodecsv

def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
````

利用该读取方法，对课程提供的三个CSV文件进行读取

观察输出的数据，都是以**字符串**的形式输出。  
这是因为 CSV库 并**不检测**各列的类型。由使用者来选择是否转换数据的类型，可以等到确实需要的时候再转换。  

例如，当需要检查 is_cancelled是否为真时，可以利用代码 `if is_concelled == 'True'`来比对字符串 **'True'**，而不一定要用布尔值

然而，比较好的做法是一开始就转换好所有的类型，以免因忘记部分类型是错误的而要花时间排除错误。

## 修正数据类型

````python
from datetime import datetime as dt

# 将字符串格式的时间转为 Python datetime 类型的时间。
# 如果没有时间字符串传入，返回 None

def parse_date(date):
    if date == '':
        return None
    else:
        return dt.strptime(date, '%Y-%m-%d')
    
# 将可能是空字符串或字符串类型的数据转为 整型 或 None。

def parse_maybe_int(i):
    if i == '':
        return None
    else:
        return int(i)

# 清理 enrollments 表格中的数据类型

for enrollment in enrollments:
    enrollment['cancel_date'] = parse_date(enrollment['cancel_date'])
    enrollment['days_to_cancel'] = parse_maybe_int(enrollment['days_to_cancel'])
    enrollment['is_canceled'] = enrollment['is_canceled'] == 'True'
    enrollment['is_udacity'] = enrollment['is_udacity'] == 'True'
    enrollment['join_date'] = parse_date(enrollment['join_date'])
    
enrollments[0]
````

````python
# 清理 engagement 的数据类型
for engagement_record in daily_engagement:
    engagement_record['lessons_completed'] = int(float(engagement_record['lessons_completed']))
    engagement_record['num_courses_visited'] = int(float(engagement_record['num_courses_visited']))
    engagement_record['projects_completed'] = int(float(engagement_record['projects_completed']))
    engagement_record['total_minutes_visited'] = float(engagement_record['total_minutes_visited'])
    engagement_record['utc_date'] = parse_date(engagement_record['utc_date'])
    
daily_engagement[0]
````

````python
# 清理 submissions 的数据类型
for submission in project_submissions:
    submission['completion_date'] = parse_date(submission['completion_date'])
    submission['creation_date'] = parse_date(submission['creation_date'])

project_submissions[0]
````

> 导师询问优达学城的数据学家后，了解到：这些学生数据都是优达学城的测试账号（因为`is_udacity`是True），不一定会在daily_engagement中有信息

````python
# 找出所有优达学城的测试账号

udacity_test_accounts = set()
for enrollment in enrollments:
    if enrollment['is_udacity']:
        udacity_test_accounts.add(enrollment['account_key'])
len(udacity_test_accounts)
````

````python
# 删除优达学城的测试账号

def remove_udacity_accounts(data):
    non_udacity_data = []
    for data_point in data:
        if data_point['account_key'] not in udacity_test_accounts:
            non_udacity_data.append(data_point)
    return non_udacity_data
````

````python
non_udacity_enrollments = remove_udacity_accounts(enrollments)
non_udacity_engagement = remove_udacity_accounts(daily_engagement)
non_udacity_submissions = remove_udacity_accounts(project_submissions)

print len(non_udacity_enrollments)
print len(non_udacity_engagement)
print len(non_udacity_submissions)
````

````python
# 创建满足下列条件的字典 paid_students：
# 1. 学生还未注销 (days_to_cancel is None)
# 2. 参与超过七天 (days_to_cancel > 7)

# keys: account keys
# Values: enrollment date

# 要记得使用 non_udacity_enrollments （去除了测试账号）
paid_students = dict()
for enrollment in non_udacity_enrollments:
    days_to_cancel = enrollment['days_to_cancel']
    if days_to_cancel is None or days_to_cancel > 7:
        account_key = enrollment['account_key']
        enrollment_date = enrollment['join_date']
        
        # 存在同一个人多次注册的情况
        # 故而应保存最近的注册日期
        if account_key not in paid_students or \
                enrollment_date > paid_students[account_key]:
            paid_students[account_key] = enrollment_date

len(paid_students)
````

## 获取第一周的数据

````python
# 删除 与免费试用期中注销的学生 对应的数据点
def remove_free_trial(data):
    new_data = []
    for data_point in data:
        if data_point['account_key'] in paid_students:
            new_data.append(data_point)
    return new_data
````

````python
paid_enrollments = remove_free_trial(non_udacity_enrollments)
paid_engagement = remove_free_trial(non_udacity_engagement)
paid_submissions = remove_free_trial(non_udacity_submissions)

print len(paid_enrollments)
print len(paid_engagement)
print len(paid_submissions)
````

````python
# 基于学生的加入日期和特定一天的互动记录
# 若该互动记录发生在学生加入1周内，则反回 True

def within_first_week(join_date, engage_date):
    # 两个datetime的差值，是datetime.timedelta
    # 不能直接跟 int 比较大小
    # 需要先用 timedelta.days 获取相差天数（int）
    time_delta = engage_date - join_date
    return time_delta.days < 7
````

````python
## 创建一个 engagement 记录的列表
## 该列表只包括付费学生以及加入的前7天的学生的记录
## 输入符合要求的行数

paid_engagement_in_first_week = []
for engagement in paid_engagement:
    account_key = engagement['account_key']
    engage_date = engagement['utc_date']
    join_date = paid_students[account_key]
    if within_first_week(join_date, engage_date):
        paid_engagement_in_first_week.append(engagement)
len(paid_engagement_in_first_week)
````

## 探索的新问题
### 学生第一周上课的平均时间
Average Minutes Spent in Classroom

1. 对参与记录进行分组：使各组分别含有某学生的所有参与记录  
使用字典来保存，将key设为`account_key`、value设为该学生的参与记录
2. 将各学生的所有参与记录数据相加，计算各学生上课的总时间
3. 计算总数的平均值

````python
from collections import defaultdict
# 创建基于 student 对 engagement 进行分组的字典
# 字典的键为帐号（account key），值为包含互动记录的列表


# 在字典中寻找不存在的关键字，会得到空列表
engagement_by_account = defaultdict(list) 
for engagement_record in paid_engagement_in_first_week:
    account_key = engagement_record['account_key']
    engagement_by_account[account_key].append(engagement_record)
    
````

````python
total_minutes_by_account = {}

for account_key, engagement_for_student in engagement_by_account.items():
    total_minutes = 0
    for engagement_record in engagement_for_student:
        total_minutes += engagement_record['total_minutes_visited']
    total_minutes_by_account[account_key] = total_minutes
````

````python
total_minutes = total_minutes_by_account.values()

import numpy as np

print 'Mean:', np.mean(total_minutes)
print 'Standard deviation:', np.std(total_minutes)
print 'Minimum:', np.min(total_minutes)
print 'Maximum:', np.max(total_minutes)
````

结果：

#### 发现的问题

一个学生可以注册多次（注册后注销然后再注册）。 

在第一次注册后，就会开始产生engagement_records。 然而我们在生成paid_students时是保留最近一次注册的信息。

故而，在within_first_week函数里，判断join_date和engage_date的差值时，没有考虑排除负值（即在最新一次注册日期之前产生的engagement_records）

修改函数`within_first_week`，重新运行相关的代码

````python
# 基于学生的加入日期和特定一天的互动记录
# 若该互动记录发生在学生加入1周内，则反回 True

def within_first_week(join_date, engage_date):
    # 两个datetime的差值，是datetime.timedelta
    # 不能直接跟 int 比较大小
    # 需要先用 timedelta.days 获取相差天数（int）
    
    # 要排除负值（即在最新一次注册日期之前产生的engagement_records）
    time_delta = engage_date - join_date
    return time_delta.days >= 0 and time_delta.days < 7
````

````python
paid_engagement_in_first_week = []
for engagement in paid_engagement:
    account_key = engagement['account_key']
    engage_date = engagement['utc_date']
    join_date = paid_students[account_key]
    if within_first_week(join_date, engage_date):
        paid_engagement_in_first_week.append(engagement)
print len(paid_engagement_in_first_week)

from collections import defaultdict

engagement_by_account = defaultdict(list) 
for engagement_record in paid_engagement_in_first_week:
    account_key = engagement_record['account_key']
    engagement_by_account[account_key].append(engagement_record)

total_minutes_by_account = {}

for account_key, engagement_for_student in engagement_by_account.items():
    total_minutes = 0
    for engagement_record in engagement_for_student:
        total_minutes += engagement_record['total_minutes_visited']
    total_minutes_by_account[account_key] = total_minutes
    
total_minutes = total_minutes_by_account.values()

import numpy as np

print 'Mean:', np.mean(total_minutes)
print 'Standard deviation:', np.std(total_minutes)
print 'Minimum:', np.min(total_minutes)
print 'Maximum:', np.max(total_minutes)
````

对于重复利用的代码，提取为函数
1. 一个用于按照key_name对记录进行分组
2. 一个用于计算各个分组的条目总数
3. 一个用来打印平均值、标准偏差等统计结果

````python
# 用于按照key_name对记录进行分组
from collections import defaultdict

# 以数据data作为参数，返回映射key_name和记录列表的字典
def group_data(data, key_name):
    grouped_data = defaultdict(list)
    for data_point in data:
        key = data_point[key_name]
        grouped_data[key].append(data_point)
    return grouped_data
````

````python
# 用于计算grouped_data各个分组的field_name条目总数

def sum_grouped_items(grouped_data, field_name):
    summed_data = {}
    
    for key, data_points in grouped_data.items():
        total = 0
        for data_point in data_points:
            total += data_point[field_name]
        summed_data[key] = total
    
    return summed_data
````

````python
# 用来打印平均值、标准偏差等统计结果
import numpy as np

def describe_data(data):
    print 'Mean:', np.mean(data)
    print 'Standard deviation:', np.std(data)
    print 'Minimum:', np.min(data)
    print 'Maximum:', np.max(data)
````

````python
lessons_by_account = group_data(paid_engagement_in_first_week, 'account_key')
total_completed_lessons = sum_grouped_items(lessons_by_account, 'lessons_completed')
describe_data(total_completed_lessons.values())
````
